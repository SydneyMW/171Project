```python
import pandas as pd
```

```python
data = pd.read_csv('./dataset/ad.data', header=None)
data.head()
```

```python
with open('./dataset/ad.names', "r", encoding="iso-8859-1") as file:
    lines = [line.rstrip() for line in file]
```

```python
unparsed = pd.DataFrame(lines)
unparsed.columns = ['Original']
unparsed[0:12]
```

```python
unparsed['Description'] = unparsed['Original'].str.contains('\|')
unparsed['Blank'] = unparsed['Original'] == ''
unparsed['Feature'] = None
unparsed['FVal'] = None
unparsed['FType'] = None
for i in range(len(unparsed)):
    if (unparsed['Description'][i] == False) & (unparsed['Blank'][i] == False):
        feature, fval = unparsed['Original'][i].split(": ")
        fval = fval[:-1]
    else:
        feature, fval = None, None
    unparsed['Feature'][i] = feature
    unparsed['FVal'][i] = fval
    if (fval == '0,1'):
        ftype = 'Binary'
    elif fval == 'Continuous':
        ftype = 'Continuous'
    else:
        ftype = 'Other'
    unparsed['FType'][i] = ftype
```

```python
unparsed
```

```python
parsed = unparsed[(unparsed.Description == False) & (unparsed.Blank == False)]
parsed = parsed.drop(columns = ['Description', 'Blank'])
parsed
```

```python
feature_names = list(parsed.Feature)
feature_names = feature_names + ['ad']
len(feature_names)
```

```python
data.shape
```

```python
combined_df = data
combined_df.columns = list(feature_names)
combined_df
```
